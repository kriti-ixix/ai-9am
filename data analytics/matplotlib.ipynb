```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

# Functional Method

```python
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('My Graph')
```

```python
plt.plot(x, y, 'g')
```

```python
plt.plot(x, y, '#FFC0CB')
```

```python
plt.plot(x, y, 'g--')
```

```python
plt.plot(x, y, 'bo-')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g--')

plt.tight_layout()
```

# Object Oriented Method

```python
canvas = plt.figure()

fig1 = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
fig2 = canvas.add_axes([0.15, 0.65, 0.2, 0.2])

fig1.plot(x, y, 'g')
fig2.plot(y, x, 'r')
```

```python
canvas = plt.figure(figsize=(12, 8))

fig1 = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
fig2 = canvas.add_axes([0.15, 0.65, 0.2, 0.2])

fig1.plot(x, y, 'g')
fig2.plot(y, x, 'r')
```
