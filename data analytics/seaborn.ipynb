```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

# Distribution Plots

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(x='total_bill', data=tips)
```

```python
sns.histplot(x='total_bill', data=tips, bins=25)
```

```python
sns.kdeplot(x='tip', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='time', data=tips)
```

```python
sns.countplot(x='smoker', data=tips, palette='rocket')
```

# Relational Plots

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hist')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
plt.figure(figsize=(8, 8))
sns.boxplot(x='day', y='total_bill', data=tips)
```
